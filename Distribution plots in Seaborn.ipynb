```python
import seaborn as sns
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

# Loading the Penguins Dataset

```python
penguins = sns.load_dataset('penguins')
penguins.head()
```

# Drawing a Simple displot

```python
sns.distplot(penguins.flipper_length_mm)
```

```python
sns.distplot(penguins.flipper_length_mm,rug = True)
```

# Removing the KDE plot

```python
sns.distplot(penguins.flipper_length_mm,kde = False)
```

```python
sns.distplot(penguins.flipper_length_mm,hist = False)
```

```python
sns.distplot(penguins.flipper_length_mm,kde = False,rug = True,hist = False)
```

```python
sns.distplot(penguins.flipper_length_mm,bins = [160,170,180,190,200,240])
```

# fitting the data to a particular distribution using norm_hist

```python
sns.distplot(penguins.flipper_length_mm,norm_hist = True)
```

# Styling rugplots

```python
sns.distplot(penguins.flipper_length_mm,rug_kws = {'color':'red','height':0.20},rug = True)
```

# Styling Histograms

```python
sns.distplot(penguins.flipper_length_mm,kde = False,hist_kws = {'color':'red','alpha':0.20,'histtype':'step'})
```

```python
sns.distplot(penguins.flipper_length_mm,kde = False,hist_kws = {'color':'red','alpha':0.60,'histtype':'step'})
```

```python
sns.distplot(penguins.flipper_length_mm,kde = False,hist_kws = {'color':'red','alpha':0.80,'histtype':'stepfilled'})
```

# Styling KDE plots

```python
sns.distplot(penguins.flipper_length_mm,kde = True,hist = False,kde_kws = {'color':'red','lw':2.5,'alpha':0.4})
```

##  Box BoxPlot

```python
sns.boxplot(penguins.bill_length_mm)
```

 # island v/s bill_lenght_mm,

```python
sns.boxplot(y = penguins.bill_length_mm,x = penguins.island)
```

```python
sns.boxplot(x = penguins.bill_length_mm,y = penguins.island)
```

# hue attribute as sex

```python
sns.boxplot(y = penguins.bill_length_mm,x = penguins.island,hue = penguins.sex)
```

### horizonta bar 

```python
sns.boxplot(x = penguins.bill_length_mm,y = penguins.island,hue = penguins.sex)
```

## ordering the bars 

```python
sns.boxplot(y = penguins.bill_length_mm,x = penguins.island,hue = penguins.sex,order = ['Biscoe','Torgersen','Dream'])
```

# how to change color attribute 

```python
sns.boxplot(y = penguins.bill_length_mm,x = penguins.island,hue = penguins.sex,order = ['Biscoe','Torgersen','Dream'],color = 'g')
```

## width property 

```python
sns.boxplot(y = penguins.bill_length_mm,x = penguins.island,hue = penguins.sex,order = ['Biscoe','Torgersen','Dream'],color = 'g',width = 1.5)
```

## whis property

```python
sns.boxplot(y = penguins.bill_length_mm,x = penguins.island,hue = penguins.sex,order = ['Biscoe','Torgersen','Dream'],whis = .5)
```

# showcaps property

```python
sns.boxplot(y = penguins.bill_length_mm,x = penguins.island,hue = penguins.sex,order = ['Biscoe','Torgersen','Dream'],showcaps = False)
```

```python
sns.boxplot(y = penguins.bill_length_mm,x = penguins.island,hue = penguins.sex,order = ['Biscoe','Torgersen','Dream'],showcaps = True)
```
